```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
```

```python
ATE_list1 = np.loadtxt("ATE_result1.csv")[:300]
ATE_list2 = np.loadtxt("ATE_result2.csv")[:300]
ATE_list3 = np.loadtxt("ATE_result3.csv")[:300]
ATE_list4 = np.loadtxt("ATE_result4.csv")[:300]
```

```python
ATE_list_list = [ATE_list1, ATE_list2, ATE_list3, ATE_list4]
```

```python
for i in range(len(ATE_list_list)):
    print(len(ATE_list_list[i]))
```

```python
true_ATE = 5
```

```python
mse_list = []
bias_list = []

for ATE_list in ATE_list_list:
    mse_list.append(np.mean((np.array(ATE_list) - true_ATE)**2, axis=0))
    bias_list.append(np.mean((np.array(ATE_list) - true_ATE), axis=0))
```

```python
results_list = []

for i in range(len(ATE_list_list)):
    mse_temp = []
    bias_temp = []
    for j in [1, 0, 2, 3, 5, 9, 11, 6, 8, 16, 17, 18, 12]:
        mse_temp.append(mse_list[i][j])
        bias_temp.append(bias_list[i][j])
        
    results_list.append(mse_temp)
    results_list.append(bias_temp)
```
